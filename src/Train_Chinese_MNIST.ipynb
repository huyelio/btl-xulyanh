```python
"""
Google Colab Script - Train Chinese MNIST v·ªõi Preprocessing Pipeline
√Åp d·ª•ng ƒê√öNG pipeline x·ª≠ l√Ω ·∫£nh nh∆∞ trong app.py ƒë·ªÉ model h·ªçc ƒë√∫ng
"""

# ========== SETUP ==========
print("=" * 60)
print("B∆Ø·ªöC 1: C√ÄI ƒê·∫∂T V√Ä T·∫¢I D·ªÆ LI·ªÜU")
print("=" * 60)

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
import os
import zipfile
from PIL import Image
from pathlib import Path
import cv2

print("‚úì ƒê√£ import th∆∞ vi·ªán")
print(f"TensorFlow: {tf.__version__}")
print(f"GPU: {tf.config.list_physical_devices('GPU')}")

# ========== KAGGLE API ==========
print("\n" + "=" * 60)
print("C√ÄI ƒê·∫∂T KAGGLE API")
print("=" * 60)
print("1. Truy c·∫≠p: https://www.kaggle.com/settings/account")
print("2. T·∫°o API token v√† t·∫£i kaggle.json")
print("3. Upload kaggle.json v√†o Colab\n")

from google.colab import files
print("üì§ Upload file kaggle.json:")
uploaded = files.upload()

!pip install -q kaggle
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

print("‚úì ƒê√£ c√†i ƒë·∫∑t Kaggle API")

# ========== T·∫¢I DATASET ==========
print("\n" + "=" * 60)
print("T·∫¢I DATASET")
print("=" * 60)

!kaggle datasets download -d gpreda/chinese-mnist

print("‚úì ƒê√£ t·∫£i dataset")

print("\nüì¶ Gi·∫£i n√©n...")
with zipfile.ZipFile('chinese-mnist.zip', 'r') as zip_ref:
    zip_ref.extractall('chinese_mnist_data')
print("‚úì ƒê√£ gi·∫£i n√©n")

# ========== PREPROCESSING FUNCTION ==========
print("\n" + "=" * 60)
print("B∆Ø·ªöC 2: ƒê·ªäNH NGHƒ®A PREPROCESSING PIPELINE")
print("=" * 60)

def preprocess_for_chinese_training(image_array: np.ndarray) -> np.ndarray:
    """
    Pipeline x·ª≠ l√Ω ·∫£nh GI·ªêNG H·ªÜT trong app.py
    Input: ·∫£nh PIL ƒë√£ convert sang numpy (H, W) ho·∫∑c (H, W, 3)
    Output: ·∫£nh ƒë√£ x·ª≠ l√Ω (64, 64) uint8
    """
    # Chuy·ªÉn sang grayscale n·∫øu c·∫ßn
    if len(image_array.shape) == 3:
        gray = cv2.cvtColor(image_array, cv2.COLOR_RGB2GRAY)
    else:
        gray = image_array.copy()

    # L√†m m·ªù ƒë·ªÉ gi·∫£m nhi·ªÖu
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Threshold Otsu
    _, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # X√°c ƒë·ªãnh background v√† invert n·∫øu c·∫ßn
    corners = [thresh[0, 0], thresh[0, -1], thresh[-1, 0], thresh[-1, -1]]
    if np.mean(corners) > 127:
        inverted = cv2.bitwise_not(thresh)
    else:
        inverted = thresh.copy()

    # T√¨m bounding box c·ªßa T·∫§T C·∫¢ pixels tr·∫Øng
    coords = cv2.findNonZero(inverted)
    if coords is not None and len(coords) > 0:
        x, y, w, h = cv2.boundingRect(coords)

        # Crop v·ªõi padding
        pad = max(8, int(min(w, h) * 0.1))
        x1 = max(0, x - pad)
        y1 = max(0, y - pad)
        x2 = min(inverted.shape[1], x + w + pad)
        y2 = min(inverted.shape[0], y + h + pad)

        cropped = inverted[y1:y2, x1:x2]
    else:
        cropped = inverted.copy()

    # Resize gi·ªØ t·ª∑ l·ªá, fit v√†o 56x56
    h, w = cropped.shape
    target_inner = 56
    scale = min(target_inner / w, target_inner / h)
    new_w = int(w * scale)
    new_h = int(h * scale)

    resized = cv2.resize(cropped, (new_w, new_h), interpolation=cv2.INTER_AREA)

    # Center v√†o canvas 64x64
    canvas = np.zeros((64, 64), dtype=np.uint8)
    top = (64 - new_h) // 2
    left = (64 - new_w) // 2
    canvas[top:top+new_h, left:left+new_w] = resized

    # L√†m m∆∞·ª£t final
    final = cv2.GaussianBlur(canvas, (3, 3), 0)

    return final

print("‚úì ƒê√£ ƒë·ªãnh nghƒ©a preprocessing pipeline (GI·ªêNG app.py)")

# ========== ƒê·ªåC V√Ä X·ª¨ L√ù D·ªÆ LI·ªÜU ==========
print("\n" + "=" * 60)
print("B∆Ø·ªöC 3: ƒê·ªåC V√Ä X·ª¨ L√ù D·ªÆ LI·ªÜU")
print("=" * 60)

# ƒê·ªçc metadata
csv_path = 'chinese_mnist_data/chinese_mnist.csv'
df = pd.read_csv(csv_path)
print(f"‚úì ƒê·ªçc CSV: {df.shape}")
print(f"\nüìä Ph√¢n b·ªë labels:")
print(df['character'].value_counts().sort_index())

# T√¨m file ·∫£nh
data_dir = Path('chinese_mnist_data/data/data')
if not data_dir.exists():
    data_dir = Path('chinese_mnist_data/data')

image_files = sorted(data_dir.glob('input_*.jpg'))
print(f"\n‚úì T√¨m th·∫•y {len(image_files)} file ·∫£nh")

# ƒê·ªçc v√† x·ª≠ l√Ω ·∫£nh
print("\nüñºÔ∏è ƒêang ƒë·ªçc v√† x·ª≠ l√Ω ·∫£nh...")
images = []
labels = []

for i, img_path in enumerate(image_files):
    filename = img_path.stem
    parts = filename.split('_')

    if len(parts) >= 4:
        try:
            code = int(parts[3])

            # ƒê·ªçc ·∫£nh g·ªëc
            img = Image.open(img_path).convert('L')
            img_array = np.array(img)

            # √ÅP D·ª§NG PREPROCESSING PIPELINE
            processed = preprocess_for_chinese_training(img_array)

            images.append(processed)
            labels.append(code - 1)  # 0-indexed

            if (i + 1) % 1000 == 0:
                print(f"  ƒê√£ x·ª≠ l√Ω {i + 1}/{len(image_files)} ·∫£nh...")
        except Exception as e:
            print(f"  ‚ö†Ô∏è L·ªói {img_path.name}: {e}")
            continue

print(f"\n‚úì ƒê√£ x·ª≠ l√Ω {len(images)} ·∫£nh")

# Chuy·ªÉn sang numpy
X = np.array(images, dtype=np.float32) / 255.0  # Normalize [0, 1]
y = np.array(labels)

print(f"\nShape X: {X.shape}")
print(f"Shape y: {y.shape}")
print(f"Labels: {y.min()} ‚Üí {y.max()}")
print(f"Pixel range: [{X.min():.3f}, {X.max():.3f}]")

# Reshape cho CNN
X = X.reshape(-1, 64, 64, 1)
y_cat = keras.utils.to_categorical(y, 15)

print(f"‚úì Final X shape: {X.shape}")
print(f"‚úì Final y shape: {y_cat.shape}")

# Chia train/val
X_train, X_val, y_train, y_val = train_test_split(
    X, y_cat, test_size=0.2, random_state=42, stratify=y
)

print(f"\n‚úì Train: {X_train.shape[0]} samples")
print(f"‚úì Val: {X_val.shape[0]} samples")

# Hi·ªÉn th·ªã ·∫£nh m·∫´u SAU KHI X·ª¨ L√ù
print("\nüì∏ Hi·ªÉn th·ªã ·∫£nh sau preprocessing:")
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i].squeeze(), cmap='gray')
    label_idx = np.argmax(y_train[i])
    char = df[df['code'] == label_idx + 1]['character'].iloc[0]
    ax.set_title(f'{label_idx}: {char}')
    ax.axis('off')
plt.suptitle('·∫¢nh ƒê√É X·ª¨ L√ù (n·ªÅn ƒëen, ch·ªØ tr·∫Øng, centered)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# ========== X√ÇY D·ª∞NG MODEL ==========
print("\n" + "=" * 60)
print("B∆Ø·ªöC 4: X√ÇY D·ª∞NG MODEL")
print("=" * 60)

model = keras.Sequential([
    # Conv Block 1
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 1), padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),

    # Conv Block 2
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),

    # Conv Block 3
    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),

    # Conv Block 4
    layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),

    # Dense layers
    layers.Flatten(),
    layers.Dropout(0.5),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(15, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("‚úì Model ƒë√£ s·∫µn s√†ng")
model.summary()

# ========== HU·∫§N LUY·ªÜN ==========
print("\n" + "=" * 60)
print("B∆Ø·ªöC 5: HU·∫§N LUY·ªÜN")
print("=" * 60)

callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    )
]

print("üöÄ B·∫Øt ƒë·∫ßu training...\n")

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=64,
    validation_data=(X_val, y_val),
    callbacks=callbacks,
    verbose=1
)

print("\n‚úÖ Ho√†n th√†nh training!")

# ========== ƒê√ÅNH GI√Å ==========
print("\n" + "=" * 60)
print("B∆Ø·ªöC 6: ƒê√ÅNH GI√Å")
print("=" * 60)

val_loss, val_acc = model.evaluate(X_val, y_val, verbose=0)
print(f"\nüìä Validation Results:")
print(f"  Loss: {val_loss:.4f}")
print(f"  Accuracy: {val_acc*100:.2f}%")

# Plot history
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.plot(history.history['accuracy'], label='Train')
ax1.plot(history.history['val_accuracy'], label='Val')
ax1.set_title('Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)

ax2.plot(history.history['loss'], label='Train')
ax2.plot(history.history['val_loss'], label='Val')
ax2.set_title('Loss')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# Predictions
print("\nüìä D·ª± ƒëo√°n m·∫´u:")
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
preds = model.predict(X_val[:10], verbose=0)

for i, ax in enumerate(axes.flat):
    ax.imshow(X_val[i].squeeze(), cmap='gray')

    true_idx = np.argmax(y_val[i])
    pred_idx = np.argmax(preds[i])
    conf = preds[i][pred_idx] * 100

    true_char = df[df['code'] == true_idx + 1]['character'].iloc[0]
    pred_char = df[df['code'] == pred_idx + 1]['character'].iloc[0]

    color = 'green' if true_idx == pred_idx else 'red'
    ax.set_title(f'True: {true_char}\nPred: {pred_char} ({conf:.1f}%)', color=color)
    ax.axis('off')

plt.tight_layout()
plt.show()

# ========== L∆ØU MODEL ==========
print("\n" + "=" * 60)
print("B∆Ø·ªöC 7: L∆ØU MODEL")
print("=" * 60)

filename = 'chinese_model.h5'
model.save(filename)
print(f"‚úì ƒê√£ l∆∞u: {filename}")

size_mb = os.path.getsize(filename) / (1024 * 1024)
print(f"üì¶ Size: {size_mb:.2f} MB")

print("\nüì• T·∫£i v·ªÅ m√°y...")
files.download(filename)

print("\n" + "=" * 60)
print("‚úÖ HO√ÄN T·∫§T!")
print("=" * 60)
print(f"""
T·ªîNG K·∫æT:
‚úì S·ªë ·∫£nh: {len(images)}
‚úì Validation Accuracy: {val_acc*100:.2f}%
‚úì Model: {filename}

QUAN TR·ªåNG:
‚úÖ Model ƒë√£ ƒë∆∞·ª£c train v·ªõi C√ôNG preprocessing pipeline nh∆∞ app.py
‚úÖ ·∫¢nh training: n·ªÅn ƒëen, ch·ªØ tr·∫Øng, centered (gi·ªëng khi predict)
‚úÖ ƒê·ªô ch√≠nh x√°c s·∫Ω cao h∆°n nhi·ªÅu so v·ªõi version tr∆∞·ªõc

B∆Ø·ªöC TI·∫æP THEO:
1. Copy {filename} v√†o: btl_final/models/
2. Ch·∫°y app.py ƒë·ªÉ test model

LABELS:
0:Èõ∂  1:‰∏Ä  2:‰∫å  3:‰∏â  4:Âõõ  5:‰∫î  6:ÂÖ≠  7:‰∏É
8:ÂÖ´  9:‰πù  10:ÂçÅ  11:Áôæ  12:ÂçÉ  13:‰∏á  14:‰∫ø
""")
```
